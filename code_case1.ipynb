TÍNH CÁC CHỈ SỐ

```python
def calculate_and_print_stats(prices, name, dates_arr):
    # Tính toán các chỉ số
    mean_val = np.mean(prices)
    std_val = np.std(prices)
    max_val = np.max(prices)
    min_val = np.min(prices)
    argmax_val = np.argmax(prices)
    argmin_val = np.argmin(prices)
    
    # In ra kết quả
    print(f"\nTHỐNG KÊ GIÁ VÀNG {name}")
    print(f"Giá Trung bình (Mean): {mean_val:,.0f}")
    print(f"Độ lệch chuẩn (Std): {std_val:,.0f}")
    print(f"Giá Cao nhất (Max): {max_val:,.0f} (Ngày: {dates_arr[argmax_val]})")
    print(f"Giá Thấp nhất (Min): {min_val:,.0f} (Ngày: {dates_arr[argmin_val]})")
    print(f"Index Max (Argmax): {argmax_val}")
    print(f"Index Min (Argmin): {argmin_val}")

# Tính cho TP.HCM và Hà Nội
calculate_and_print_stats(prices_hcm, "TP.HCM", dates)
calculate_and_print_stats(prices_hn, "HÀ NỘI", dates)
```
